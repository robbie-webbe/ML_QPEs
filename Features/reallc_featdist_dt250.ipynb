```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

feats_data = np.loadtxt('realobs_test_data_dt250.csv',delimiter=',',dtype=object)
feats = pd.read_csv('realobs_test_data_dt250.csv')
for i in range(len(feats)):
    feats.iloc[i,0] = feats_data[i+1,0]
nqpe_feats = feats[feats['QPE?']==0.0]
qpe_feats = feats[feats['QPE?']==1.0]
```
